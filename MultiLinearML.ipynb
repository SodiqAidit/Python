```python
# Supervice Machine Learning
# Multiple linear regression
# Third Project --- Predicting profit base on
# https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv


import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')

data.head(5)
```

```python
data.info()
```

```python
X = ['R&D Spend','Administration','Marketing Spend','State']
X = data[X].values
X.shape
```

```python
Y = data['Profit'].values
Y
```

```python
# OneHotEncoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers = [('OneHot',OneHotEncoder(),[3])],  # [3] was for no of index number for the column
    remainder = 'passthrough'  # please do not remove other columns which is not specified
    )

X = transformer.fit_transform(X)
X = X.astype('int')
```

```python
X
```

```python
# 0.2 mean 20% of data used for Testing
# 80% of data will be used for Training

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print('Original Shape')
print(X.shape)
print(Y.shape)

print('Training Shape')
print(x_train.shape)
print(y_train.shape)

print('Testing Shape')
print(x_test.shape)
print(y_test.shape)
```

```python
import sklearn.linear_model as lm

mlr = lm.LinearRegression()
mlr.fit(x_train,y_train)
```

```python
import pickle

# dump() is used to save ML Model
# pickle used bytestream format
pickle.dump(mlr,open('mlr.pkl','wb'))   # w --> write a file / create a file b ---> byteStream
```

```python
pred_values = mlr.predict(x_test)
print(pred_values)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,pred_values)
print('Regression Score Function :',accuracy)
print(f'Regression Score Function : {int(accuracy*100)}%')
```
